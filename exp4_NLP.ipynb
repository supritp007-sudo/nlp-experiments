First, let's define a regular expression pattern to identify URLs. Then, we'll apply this pattern to the `twitts` column to extract all URLs. Finally, we'll count the unique URLs found.

```python
# Define a regex pattern for URLs
url_pattern = r'https?://\S+'

# Extract all URLs from the 'twitts' column
all_urls = data['twitts'].apply(lambda x: re.findall(url_pattern, x))

# Flatten the list of lists into a single list of URLs
flattened_urls = [url for sublist in all_urls for url in sublist]

# Count the unique URLs
unique_url_counts = pd.Series(flattened_urls).value_counts()

print("Count of unique URLs:")
display(unique_url_counts)
```

```python
import re

pattern = r'\b[A-Za-z._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

data['email'] = data['twitts'].apply(
    lambda x: ",".join(re.findall(pattern, x))
)

data['email_count'] = data['twitts'].apply(
    lambda x: len(re.findall(pattern, x))
)
```

```python
url_pattern = r'http\S+|www\S+'
data['urls'] = data['twitts'].apply(lambda x: ",".join(re.findall(url_pattern, str(x))))
data['url_count'] = data['urls'].apply( lambda x: x.count(',') + 1 if x != '' else 0)
data
```
