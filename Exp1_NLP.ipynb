```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python


input = "hi i am akash, HOW Are you"
output = word_tokenize(input)
print(output)
```

```python
output2 = sent_tokenize(input)
print(output2)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "This is the second session"
doc =nlp(input)
```

```python
for token in doc :
  print (token)
```

```python
for sent in doc.sents :
  print(sent)
```
